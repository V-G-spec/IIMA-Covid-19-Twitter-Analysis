```python
import pandas as pd
```

```python
path = "collected.json"
df = pd.read_json(path)
df = df[df['lang'] == 'en']
df
```

```python
from preprocessing_custom import full_preprocess
df['preprocessed_text'] = [full_preprocess(text) for text in df['full_text']]
```

```python
df = df.reset_index()
df.head()
```

```python
import nltk
from nltk import word_tokenize, FreqDist
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download
nltk.download('wordnet')
nltk.download('stopwords')
from nltk.tokenize import TweetTokenizer
```

```python
lemmatizer = nltk.stem.WordNetLemmatizer()
w_tokenizer = TweetTokenizer()
def lemmatize_text(text):
    text.replace("'", " ")
    tokens_lemmed = []
    for w in w_tokenizer.tokenize(text):
        if len(w) > 2:
            tokens_lemmed.append((lemmatizer.lemmatize(w)))
    return tokens_lemmed
```

```python
stop_words = set(stopwords.words('english'))
df['tokens'] = df['preprocessed_text'].apply(lambda x: [item for item in lemmatize_text(x) if item not in stop_words])
df['tokens']
```

```python
df.to_json("proctoken.json")
```

```python
# !pip install wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
slist = []
for item in df['tokens']:
    slist.extend(item)

#Frequency of words
fdist = FreqDist(slist)
#WordCloud
wc = WordCloud(width=800, height=400, max_words=50).generate_from_frequencies(fdist)
plt.figure(figsize=(12,10))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
